```python
import pandas as pd
df=pd.read_csv("Bike_sharing_dataset.csv")
```

```python
a = df.groupby("holiday").mean()
b = df.groupby("holiday").std()

print(a['cnt'])

print(b['cnt'])
```

```python
wkm = df.groupby("weekday").mean()
wks = df.groupby("weekday").std()

print(wkm['cnt'])

print(wks['cnt'])
```

```python
wdm = df.groupby("workingday").mean()
wds = df.groupby("workingday").std()

print(wdm['cnt'])

print(wds['cnt'])
```

```python
weatherm = df.groupby("weathersit").mean()
weathers = df.groupby("weathersit").std()

print(weatherm['cnt'])

print(weathers['cnt'])
```

a.	Most trends are expected, such as holiday or weather affect the bike usage. It is a bit strange that whether it is a workday does not affect the bike usage. 

```python
from sklearn.linear_model import LinearRegression

X = df.iloc[:, 10].values.reshape(-1, 1) 
y = df.iloc[:, 15].values.reshape(-1, 1) 
```

```python
linear_regressor = LinearRegression()  
linear_regressor.fit(X, y) 
Y_pred = linear_regressor.predict(X)  

```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.plot(X, Y_pred, color='red')
plt.show()
```

```python
seasonm = df.groupby("season").mean()
seasons = df.groupby("season").std()

print(seasonm['cnt'])

print(seasons['cnt'])
```

b. The data show that rental bike has most usage during summer and significantly less usage during winter, which is expected
